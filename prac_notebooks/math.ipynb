## Math Basics for Algorithms


Write a program that reads an integer from the standard input and then prints each digit of the number on a new line in reverse order.

Example 1:

Input:

1234

Output:

4

3

2

1


```python
n=1234
while n!=0:
    ld = n%10
    n=n//10
    print(ld)
```

Count the Digits

Example 1:

Input:

1234

Output:

4





```python
n = 12345
count = 0
while n!=0:
    count+=1
    n=n//10
print(count)
```
